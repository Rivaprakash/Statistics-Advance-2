1. What is hypothesis testing in statistics?

Ans- Hypothesis testing is a way to check if a claim about a dataset is true. It uses sample data to decide whether to accept or reject a hypothesis based on probability.


2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

Ans- Null Hypothesis (H‚ÇÄ) assumes no effect or no difference (status quo).

* Alternative Hypothesis (H‚ÇÅ): Claims there is an effect or difference.

* Hypothesis testing checks if we have enough evidence to reject H‚ÇÄ in favor of H‚ÇÅ.

3. What is the significance level in hypothesis testing, and why is it important?

Ans- The significance level (Œ±) is the probability of rejecting the null hypothesis (H‚ÇÄ) when it is actually true. Common values are 0.05 or 0.01. It helps control the risk of making a Type I error (false positive).

4. What does a P-value represent in hypothesis testing?

Ans- The P-value shows the probability of getting the observed results if the null hypothesis (H‚ÇÄ) is true.

* Low P-value (‚â§ Œ±) ‚Üí Reject H‚ÇÄ (strong evidence against it).

* High P-value (> Œ±) ‚Üí Fail to reject H‚ÇÄ (not enough evidence).

It helps decide whether a result is statistically significant.

5. How do you interpret the P-value in hypothesis testing?

Ans- The P-value shows how likely the result is due to chance. A small P-value (‚â§ 0.05) means strong evidence against H‚ÇÄ, so we reject it. A large P-value (> 0.05) means not enough evidence, so we keep H‚ÇÄ.

6. What are Type 1 and Type 2 errors in hypothesis testing.

Ans- In hypothesis testing, a Type 1 error occurs when we wrongly reject a true null hypothesis (H‚ÇÄ), leading to a false positive.

A Type 2 error happens when we fail to reject a false H‚ÇÄ, leading to a false negative. Lowering one often increases the other.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

Ans- A one-tailed test checks if a value is only greater or only smaller than a given value.

A two-tailed test checks if a value is either greater or smaller than a given value.

Use a one-tailed test when you expect a specific direction.

Use a two-tailed test when you just want to check for any difference.

8. What is the Z-test, and when is it used in hypothesis testing?

When to use:

* To compare two sample means when population variances are known.

* When data follows a normal distribution or the sample is large.

```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

from scipy.stats import norm
import numpy as np

def z_test(sample, pop_mean, pop_std):
    z = (np.mean(sample) - pop_mean) / (pop_std / np.sqrt(len(sample)))
    p = 2 * (1 - norm.cdf(abs(z)))
    return z, p

sample = [52, 55, 53, 50, 49, 51, 54, 56, 48, 47]
z, p = z_test(sample, 50, 5)
print(f"Z-score: {z:.4f}, P-value: {p:.4f}")
print("Reject H‚ÇÄ" if p < 0.05 else "Fail to reject H‚ÇÄ")
```

```python
#4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

np.random.seed(42)
sample = np.random.normal(50, 5, 30)
z = (np.mean(sample) - 50) / (5 / np.sqrt(len(sample)))
p = 2 * (1 - norm.cdf(abs(z)))

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)
cv = norm.ppf(0.975)
plt.plot(x, y)
plt.axvline(z, color='r', linestyle='dashed', label=f'Z={z:.2f}')
plt.axvline(cv, color='g', linestyle='dashed', label='¬±1.96')
plt.axvline(-cv, color='g', linestyle='dashed')
plt.fill_between(x, y, where=(x > cv) | (x < -cv), color='red', alpha=0.3)
plt.legend()
plt.show()

print(f"Z-score: {z:.4f}, P-value: {p:.4f}")
print("Reject H‚ÇÄ" if p < 0.05 else "Fail to reject H‚ÇÄ")
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(alpha=0.05, beta=0.2):
    x = np.linspace(-4, 4, 1000)
    plt.plot(x, norm.pdf(x), 'b', label='H‚ÇÄ')
    plt.plot(x, norm.pdf(x, loc=2), 'r', label='H‚ÇÅ')
    cv = norm.ppf(1 - alpha)
    plt.fill_between(x, norm.pdf(x), where=(x > cv), color='blue', alpha=0.3, label='Type 1 Error (Œ±)')
    plt.fill_between(x, norm.pdf(x, loc=2), where=(x < cv), color='red', alpha=0.3, label='Type 2 Error (Œ≤)')
    plt.axvline(cv, color='black', linestyle='dashed', label='Critical Value')
    plt.legend(), plt.title("Type 1 & Type 2 Errors"), plt.show()

visualize_errors()
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
from scipy.stats import norm

def confidence_interval(sample, confidence=0.95):
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(len(sample))
    margin = norm.ppf((1 + confidence) / 2) * std_err
    return mean - margin, mean + margin

# Simulate data
np.random.seed(42)
sample = np.random.normal(50, 5, 30)

# Calculate confidence interval
ci_low, ci_high = confidence_interval(sample)
print(f"95% Confidence Interval: ({ci_low:.2f}, {ci_high:.2f})")
```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data
import numpy as np
from scipy.stats import norm

def margin_of_error(sample, confidence=0.95):
    std_err = np.std(sample, ddof=1) / np.sqrt(len(sample))
    return norm.ppf((1 + confidence) / 2) * std_err

np.random.seed(42)
sample = np.random.normal(50, 5, 30)

# Calculate margin of error
moe = margin_of_error(sample)
print(f"Margin of Error (95% confidence): {moe:.2f}")
```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

def bayes_theorem(prior, likelihood, evidence):
    posterior = (likelihood * prior) / evidence
    return posterior

prior = 0.01  # Prior probability of having a rare disease
likelihood = 0.95  # Probability of testing positive given the disease
evidence = (likelihood * prior) + ((1 - prior) * 0.05)  # Total probability of testing positive

posterior = bayes_theorem(prior, likelihood, evidence)
print(f"Posterior Probability: {posterior:.4f}")
```

```python
#16. Implement an F-test using Python to compare the variances of two random samples
import numpy as np
from scipy.stats import f

# Generate two random samples
np.random.seed(42)
sample1 = np.random.normal(50, 10, 30)
sample2 = np.random.normal(50, 15, 30)

# Calculate sample variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Compute F-statistic
f_stat = var1 / var2 if var1 > var2 else var2 / var1

df1 = len(sample1) - 1
df2 = len(sample2) - 1
p_value = 1 - f.cdf(f_stat, df1, df2)

print(f"F-test: F-statistic = {f_stat:.4f}, P-value = {p_value:.4f}")
print("Reject H‚ÇÄ" if p_value < 0.05 else "Fail to reject H‚ÇÄ")
```

```python
#20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results
import numpy as np
import pandas as pd
from statsmodels.formula.api import ols
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(42)
data = pd.DataFrame({
    'A': np.tile(['A1', 'A2'], 30),
    'B': np.repeat(['B1', 'B2', 'B3'], 20),
    'Y': np.random.normal(50, 10, 60)
})

# Two-way ANOVA
model = ols('Y ~ C(A) * C(B)', data).fit()
print(sm.stats.anova_lm(model, typ=2))

# Interaction plot
sns.pointplot(x='A', y='Y', hue='B', data=data)
plt.show()
```

```python
#22. Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy.stats import f_oneway

# Generate random samples
np.random.seed(42)
groups = [np.random.normal(50 + i * 5, 10, 30) for i in range(3)]

# ANOVA test
f_stat, p_value = f_oneway(*groups)
print(f"F = {f_stat:.4f}, P = {p_value:.4f}")

# Boxplot visualization
sns.boxplot(data=groups)
plt.xticks(ticks=[0, 1, 2], labels=['Group 1', 'Group 2', 'Group 3'])
plt.title("One-Way ANOVA")
plt.grid()
plt.show()
```

```python
#23. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy.stats import ttest_ind

# Generate random normal data
np.random.seed(42)
group1 = np.random.normal(50, 10, 30)
group2 = np.random.normal(55, 10, 30)

# Perform hypothesis testing (t-test)
t_stat, p_value = ttest_ind(group1, group2)
print(f"T = {t_stat:.4f}, P = {p_value:.4f}")

# Boxplot visualization
sns.boxplot(data=[group1, group2])
plt.xticks(ticks=[0, 1], labels=['Group 1', 'Group 2'])
plt.title("Hypothesis Testing: Mean Comparison")
plt.grid()
plt.show()
```

```python
#24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Generate random normal data
np.random.seed(42)
data = np.random.normal(50, 10, 30)

# Hypothesis test for variance
sample_var = np.var(data, ddof=1)  # Sample variance
pop_var = 100  # Hypothesized population variance
chi_sq_stat = (len(data) - 1) * sample_var / pop_var
p_value = 1 - stats.chi2.cdf(chi_sq_stat, df=len(data) - 1)

print(f"Chi-Square = {chi_sq_stat:.4f}, P = {p_value:.4f}")
print("Reject H‚ÇÄ: Variance differs from population variance" if p_value < 0.05 else "Fail to reject H‚ÇÄ: No significant difference")

# Histogram visualization
sns.histplot(data, bins=10, kde=True)
plt.title("Population Variance Hypothesis Test")
plt.grid()
plt.show()
```

```python
#25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups
import numpy as np
import scipy.stats as stats

# Sample data (successes and total observations)
success1, n1 = 45, 100  # Group 1
success2, n2 = 30, 100  # Group 2

# Proportions
p1, p2 = success1 / n1, success2 / n2
p_combined = (success1 + success2) / (n1 + n2)

# Z-test for proportions
se = np.sqrt(p_combined * (1 - p_combined) * (1/n1 + 1/n2))
z_stat = (p1 - p2) / se
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))  # Two-tailed test

print(f"Z = {z_stat:.4f}, P = {p_value:.4f}")
print("Reject H‚ÇÄ: Proportions are significantly different" if p_value < 0.05 else "Fail to reject H‚ÇÄ: No significant difference")
```

```python
#26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(42)
group1 = np.random.normal(50, 10, 30)
group2 = np.random.normal(55, 15, 30)

# Compute variances
var1, var2 = np.var(group1, ddof=1), np.var(group2, ddof=1)

# Perform F-test
f_stat = var1 / var2
df1, df2 = len(group1) - 1, len(group2) - 1
p_value = 2 * min(stats.f.cdf(f_stat, df1, df2), 1 - stats.f.cdf(f_stat, df1, df2))

print(f"F = {f_stat:.4f}, P = {p_value:.4f}")
print("Reject H‚ÇÄ: Variances are significantly different" if p_value < 0.05 else "Fail to reject H‚ÇÄ: No significant difference")

plt.boxplot([group1, group2], labels=['Group 1', 'Group 2'])
plt.title("F-Test: Variance Comparison")
plt.grid()
plt.show()
```

```python
#27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulated observed data
observed = np.array([50, 30, 20])
expected = np.array([40, 40, 20])  # Hypothetical expected distribution

# Perform Chi-square goodness of fit test
chi_sq_stat, p_value = stats.chisquare(observed, expected)

print(f"Chi-Square = {chi_sq_stat:.4f}, P = {p_value:.4f}")
print("Reject H‚ÇÄ: Observed distribution significantly differs from expected" if p_value < 0.05 else "Fail to reject H‚ÇÄ: No significant difference")

labels = ['Category 1', 'Category 2', 'Category 3']
x = np.arange(len(labels))
width = 0.4

plt.bar(x - width/2, observed, width, label='Observed', alpha=0.7)
plt.bar(x + width/2, expected, width, label='Expected', alpha=0.7)
plt.xticks(ticks=x, labels=labels)
plt.title("Chi-Square Goodness of Fit Test")
plt.legend()
plt.grid()
plt.show()
```
